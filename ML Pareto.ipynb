```python
import matplotlib.pyplot as plt
```

```python
FPR = [.3603, .2727, .2801]
FNR = [.1793, .1297, .0978]

labels = ["Decision Tree", "Random Forest", "SVM"]

par_FPR = [.2727, .2801]
par_FNR = [.1297, .0978]
```

```python
plt.scatter(FPR, FNR, color='b')
for i in range(len(labels)):
    plt.annotate(labels[i], (FPR[i], FNR[i]))
plt.plot(par_FPR, par_FNR, color='r', drawstyle='steps-post')
plt.xlabel("False Positive")
plt.ylabel("False Negative")
plt.title("Pareto Front for ML Algorithms")
ax = plt.gca()
ax.set(xlim=(0, 1), ylim=(0, 1))
plt.show()
```
