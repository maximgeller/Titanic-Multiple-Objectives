```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
FPR = [.3603, .2727, .2801, .1466, .4285, .2743]
FNR = [.1793, .1297, .0978, .2885, .2436, .1005]

labels = ["Decision Tree", "Random Forest", "SVM", "Logistic Regression", "KNN", "NN"]

par_FPR = [0, .1466, .2727, .2743, .2801, 1]
par_FNR = [1, .2885, .1297, .1005, .0978, 0]
```

```python
plt.scatter(FPR, FNR, color='b')
# for i in range(len(labels)):
#     plt.annotate(labels[i], (FPR[i], FNR[i]))
    
plt.annotate("Logistic Regression", (.1466, .31))
plt.annotate("Random Forest", (.285, .1297))
plt.annotate("NN", (.205, .11))
plt.annotate("SVM", (.295, .055))
plt.annotate("Decision Tree", (.366, .1793))
plt.annotate("KNN", (.44, .2436))

plt.scatter(par_FPR, par_FNR, color='r')
plt.plot(par_FPR, par_FNR, color='r', drawstyle='steps-post')
plt.xlabel("False Positive")
plt.ylabel("False Negative")
plt.title("Pareto Front for ML Algorithms")
ax = plt.gca()
ax.set(xlim=(0, 1.1), ylim=(0, 1.1))
plt.show()
print("Area Under Curve: %s" % (np.sum(np.abs(np.diff(par_FPR))*par_FNR[:-1])))
```
