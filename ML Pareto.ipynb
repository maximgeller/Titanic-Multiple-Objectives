```python
import matplotlib.pyplot as plt
```

```python
FPN = [.3603, .2727]
FNR = [.1793, .1297]
labels = ["Decision Tree", "Random Forest"]

par_FPN = [.2727]
par_FNR = [.1297]
```

```python
plt.scatter(FPN, FNR, color='b')
for i in range(len(labels)):
    plt.annotate(labels[i], (FPN[i], FNR[i]))
plt.plot(par_FPN, par_FNR, color='r', drawstyle='steps-post')
plt.xlabel("False Positive")
plt.ylabel("False Negative")
plt.title("Pareto Front for ML Algorithms")
ax = plt.gca()
ax.set(xlim=(0, 1), ylim=(0, 1))
plt.show()
```
