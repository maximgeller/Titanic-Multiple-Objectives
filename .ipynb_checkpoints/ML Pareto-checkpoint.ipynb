```python
import matplotlib.pyplot as plt
```

```python
FPN = [.3603, .2727, .2801, 0.1466]
FNR = [.1793, .1297, .0978, 0.2885]

labels = ["Decision Tree", "Random Forest", "SVM", "KNeighborsClassifier"]

par_FPN = [.2801]
par_FNR = [.0978]
```

```python
plt.scatter(FPN, FNR, color='b')
for i in range(len(labels)):
    plt.annotate(labels[i], (FPN[i], FNR[i]))
plt.plot(par_FPN, par_FNR, color='r', drawstyle='steps-post')
plt.xlabel("False Positive")
plt.ylabel("False Negative")
plt.title("Pareto Front for ML Algorithms")
ax = plt.gca()
ax.set(xlim=(0, 1), ylim=(0, 1))
plt.show()
```
